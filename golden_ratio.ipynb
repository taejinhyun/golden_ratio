```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.collections import PatchCollection
import matplotlib.patches as mpatches
```

##  function golden_ratio  needs 3 parms 

* number of discs : int
* number of colors : int
* angle : float

```python
def golden_ratio(num_discs,num_colors,angle): # 문제 해결을 위한 함수를 생성합니다.
    ox=0.5; oy=0.4 
    ratio=(1+5**0.5)/2.0 # 비율을 구하는 수식입니다.
    inc = (2-ratio)*2*np.pi + float(angle) # increase의 약자로 다음 원의 위치를 구하는 수식으로 예상됩니다..
    
    theta=0 #원의 초기 값입니다.
    
    k=0.1 # scale factor
    
    drad=k*(1+5**0.5)/4.0 # 반지름을 구하는 수식입니다. 
    
    minv=0 #최소값의 초기 설정입니다.
    maxv=0 #최대값의 초기 설정입니다.
    
    # minv and maxv will be used later to display inputs chosen

    # now collect in list 'patches' the locations of all the discs
    
    patches = []
    for j in range(1,num_discs+1): #1부터 지정한 원의개수까지 반복문을 실행합니다. j=1,2,3,4...num_discs까지 대입됩니다.
        r = k*j**0.5 # 
        theta += inc # 원에 다음 위치를 구합니다
        x = ox + r*np.cos(theta) # 
        y = oy + r*np.sin(theta) #
        
        # y의 최대값, 최소값을 계속 확인하여 갱신합니다.
        
        if y > maxv: # 
            maxv=y 
        elif y < minv:
            minv=y
        
        #원을 disc에 저장하고 patches list에 추가합니다.
        disc = mpatches.Circle((x,y),drad) 
        patches.append(disc)

    # start building the plot 
    fig = plt.figure()
    ax = plt.axes([0,0,1,1])

    maxv=maxv*0.95 # 

    # build colour cycle, using a number between 0 and 100 for each colour
    # color를 지정하는 부분인데 어떤 방식인지는 잘 모르겠습니다.. ㅠ 
    colcycle=[]
    s=100/num_colors # 
    for j in range(num_discs): #
        colcycle.append((j%num_colors)*s) # 

    # bring together the information for locations and colours of discs
    collection = PatchCollection(patches, cmap=matplotlib.cm.jet, alpha=1.0)
    collection.set_array(np.array(colcycle))
    ax.add_collection(collection)

    ax.set_xticks([]); ax.set_yticks([]) # suppress display of axes
    plt.axis('equal')
    plt.show() # display the plot we have built
```

```python
golden_ratio(200,20,45)
```
